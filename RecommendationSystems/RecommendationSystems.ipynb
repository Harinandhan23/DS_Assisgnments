## EDA

```python
df = df.rename(columns = {'User.ID' : 'UserID', 'Book.Title' : 'BookTitle', 'Book.Rating' : 'BookRating'})
df.head(5)
```

```python
df = df.drop(['Unnamed: 0'], axis = 1)
```

```python
df.sort_values(['UserID'])
```

```python
new_df = df.pivot_table(index = 'UserID', columns = 'BookTitle', values = 'BookRating').reset_index(drop = True)
```

```python
new_df.shape
```

```python
new_df.index = df.UserID.unique()
```

```python
new_df
```

```python
new_df.fillna(0, inplace = True)
new_df
```

## Calculating similarity

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation

book_sim = 1 - pairwise_distances(new_df.values, metric = 'cosine')
```

```python
book_sim
```

```python
book_sim_df = pd.DataFrame(book_sim)
```

```python
book_sim_df.index = df['UserID'].unique()
book_sim_df.columns = df['UserID'].unique()
book_sim_df
```

```python
import numpy as np

np.fill_diagonal(book_sim_df.values,0)
book_sim_df
```
